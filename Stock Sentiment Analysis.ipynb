```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv', encoding ='ISO-8859-1')
df.head()
```

```python
train_df = df[df['Date']<'20150101']
test_df = df[df['Date']> '20141231']
```

```python
#X = df.drop('Label', axis =1)
#Y = df['Label']
```

```python
train_df.head()
```

```python
data = train_df.iloc[:,2:27]
data.replace("[^a-zA-Z]", ' ', regex = True, inplace = True)
data.head()
```

```python
train_df['Label'][:5]
```

```python
list1 = [ i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head()
```

```python
for idex in data:
    data[idex] = data[idex].str.lower()
data.head()    
```

```python
headlines =[]
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, 0:25]))
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
count = CountVectorizer(ngram_range =(2,2))
traindataset= count.fit_transform(headlines)
```

```python
#print(traindataset)
```

```python
randomclass = RandomForestClassifier(n_estimators=100,criterion='gini')
randomclass.fit(traindataset, train_df['Label'])
```

```python
randomclass = RandomForestClassifier(n_estimators=300,criterion='gini')
randomclass.fit(traindataset, train_df['Label'])
```

```python
test_transform = []
for row in range(0, len(test_df.index)):
    test_transform.append(' '.join(str(x) for x in test_df.iloc[row, 2:27]))
test_dataset = count.transform(test_transform)
predictions = randomclass.predict(test_dataset)
```

```python
print(test_dataset)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test_df['Label'], predictions)
```

```python
score = accuracy_score(test_df['Label'], predictions)
```

```python
report = classification_report(test_df['Label'], predictions)
```

```python
Final = pd.DataFrame({'Matrix_Score' : ['Confusion_Matrix' , 'Accuracy_Score', 'Report'], 'Score' : [matrix, score, report]})
```
